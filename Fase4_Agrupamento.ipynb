## Dados (Definição e Tratamentos)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
df = pd.read_csv("https://raw.githubusercontent.com/erlonL/Projeto-ICD/main/heart-disease.csv")
df
```

```python
df = df.dropna() # retirando possíveis elementos NaN/Na
```

```python
# Tratamento para melhor visualização
def troca_por_categoria(valor_coluna):
  if valor_coluna == "Poor":
    valor_coluna = 1
  elif valor_coluna == "Fair":
    valor_coluna = 2
  elif valor_coluna == "Good":
    valor_coluna = 3
  elif valor_coluna == "Very good":
    valor_coluna = 4
  elif valor_coluna == "Excellent":
    valor_coluna = 5
  return valor_coluna

df['GenHealthNUM'] = df['GenHealth'].apply(troca_por_categoria)
```

# Agrupamento

### Gráficos de dispersão

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df[['BMI','SleepTime']])
sns.scatterplot(data = df, x='BMI',y='SleepTime', hue = kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df[['BMI','PhysicalHealth']])
sns.scatterplot(data = df, x = "BMI", y = "PhysicalHealth", hue = kmeans.labels_)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df[['BMI','MentalHealth']])
sns.scatterplot(data = df, x = "BMI", y = "MentalHealth", hue=kmeans.labels_)
plt.show()
```
